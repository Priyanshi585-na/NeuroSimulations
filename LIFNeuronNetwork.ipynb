```python
import numpy as np
import matplotlib.pyplot as plt

V_rest = -70
V = -70
V_th = -55
R = 3
T = 10
I = 10
dt = 0.1

t = np.arange(0,50,dt)
v = np.zeros(len(t))
v[0] = V

for i in range(1,len(t)-1):
    v[i] = v[i-1] + (dt/T)*(-(v[i-1]-V_rest) + R*I)
    if v[i]>=V_th:
        v[i] = 50
        v[i] = V
    
plt.figure(figsize = (10,5))
plt.plot(t,v)
plt.xlabel("time(ms)")
plt.ylabel("membrane voltage(mV)")
plt.title("LIF neuron network")
plt.axhline(V_th, color='r', linestyle='--', label='Threshold')
plt.grid()
plt.show()
```
